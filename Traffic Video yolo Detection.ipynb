```python
import cv2

video_path = r"C:\Users\ashwi\OneDrive\Desktop\YOLO\projects\Traffic vehicle\runs\detect\track\sample.avi"

cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow("YOLO Detection", frame)

    if cv2.waitKey(25) & 0xFF == ord("q"):  # Press 'q' to exit
        break

cap.release()
cv2.destroyAllWindows()
```
